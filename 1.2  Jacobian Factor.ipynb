```python
s_ = s + 100
print ( np.all(s_ >= -1), np.all(s_ < 0) )
count, bins, ignored = plt.hist(s_, 10, density=True)
# Shift in origin of the x
```

```python
s_ = s * 2
print ( np.all(s_ >= -1), np.all(s_ < 0) )
count, bins, ignored = plt.hist(s_, 10, density=True)
```

##### Non-Linear Transformation

```python
s_ = s * s
print ( np.all(s_ >= -1), np.all(s_ < 0) )
count, bins, ignored = plt.hist(s_, 10, density=True)
```

```python
s_ = s * s
weights = np.ones_like(s_)/float(len(np.array(s_)))
_ = plt.hist(s_, weights=weights, bins = 10)
```
