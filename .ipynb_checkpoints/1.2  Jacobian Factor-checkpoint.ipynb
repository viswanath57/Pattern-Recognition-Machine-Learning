```python
import pandas as pd
import math
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt

import random

from collections import Counter
```

```python
s = np.random.uniform(-1,0,100000)
print ( np.all(s >= -1), np.all(s < 0) )
count, bins, ignored = plt.hist(s, 10, density=True)
```

```python
s_ = s + 100
print ( np.all(s_ >= -1), np.all(s_ < 0) )
count, bins, ignored = plt.hist(s_, 10, density=True)
# Shift in origin of the x
```

```python
s_ = s * 2
print ( np.all(s_ >= -1), np.all(s_ < 0) )
count, bins, ignored = plt.hist(s_, 100, density=True)
```

```python
s_ = s * s
print ( np.all(s_ >= -1), np.all(s_ < 0) )
count, bins, ignored = plt.hist(s_, 100, density=True)
```

```python
s_ = 0
for i in range(100):
    s_ = s_ + count[i] * 1/100
    print ( f'{count[i]*0.01:.5f}' )
s_    
```

Lesson Learnt: Linear transformation does not change area under one rectangle, but that rectangle can get contract or expand

```python
x = np.random.uniform(0,1,100000)
weights = np.ones_like(x)/float(len(np.array(x)))
vec_ = plt.hist(x, weights=weights, bins = 100)
```

```python
x_ = x +10
weights = np.ones_like(x_)/float(len(np.array(x_)))
vec_ = plt.hist(x_, weights=weights, bins = 100)
```

```python
x_ = 2 * x
weights = np.ones_like(x_)/float(len(np.array(x_)))
vec_ = plt.hist(x_, weights=weights, bins = 100)
```
